## Graph of Tweet Text Length by Percentage of All Tweets for a Given Company:


The code snippet provides character counts for all the Tweets associated with a given company.  We visualize this statistic by plotting a relative frequency histogram of those counts for the CSV dataset.<br>



```python
    print("Character Count Statistics of Tweet Text by Associated Company: ")
    print("Character Count Relative Frequency Histogram: ")
    plt.figure()
    grid = sns.FacetGrid(tweet_dataframe[['text_derived', 'company_derived_designation']],
                         col='company_derived_designation', col_wrap=6, ylim=(0, 1))
    grid.map_dataframe(tweet_util_v2.relhist_proc, 'text_derived', bins=10, proc=tweet_util_v2.char_len)
    grid.set_titles('{col_name}')
    grid.set_xlabels("# of Characters").set_ylabels("Percentage of All Tweets")
    plt.show()
```


Y-axis labels indicate the percentage of all Tweets for the given company that are of that character length.  X-axis labels provide the number of character each bar represents.<br>

It seems that most Tweets are around 200-300 characters long, though the graphs aren't exactly that precise.  Most Tweets for a given company share the same length, with the outliers tapering off to fewer or more characters.<br>
